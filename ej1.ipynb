**2\. Composición:**

**¿Qué representan las instancias que componen el conjunto de datos (por ejemplo, documentos, fotos, personas, países)?**   
**¿Hay varios tipos de instancias (por ejemplo, películas, usuarios y calificaciones; personas e interacciones entre ellas; nodos y bordes)? Proporcione una descripción.**

Cada instancia representa un contacto telefónico individual (una llamada) realizado a un cliente en el marco de una campaña de marketing. No existen diversos tipos de instancias ni estructuras complejas de nodos o bordes; se trata de un archivo donde cada fila consolida tanto el perfil demográfico del cliente como el contexto de la interacción y las variables macroeconómicas del momento. Según la documentación de la UCI, cada registro funciona como una unidad atómica que vincula al sujeto con el resultado final de la comunicación.


**3\. Proceso de recopilación:**

**¿Cómo se adquirieron los datos asociados con cada instancia? ¿Los datos fueron observados directamente (por ejemplo, texto sin procesar, calificaciones de películas), informados por los sujetos (por ejemplo, respuestas de encuestas) o indirectamente inferidos/derivados de otros datos (por ejemplo, etiquetas de partes del discurso, conjeturas basadas en modelos para edad o idioma)? Si los datos fueron informados por los sujetos o indirectamente inferidos/derivados de otros datos, ¿se validaron/verificaron? Si es así, describa cómo.**

La recopilación fue un proceso híbrido basado principalmente en la observación directa de registros administrativos de una institución bancaria portuguesa entre 2008 y 2010\. No se trató de una encuesta, sino de un relevamiento de datos reales alojados en los sistemas del banco.  
Si bien hay datos demográficos (como el trabajo o nivel educativo) que fueron informados originalmente por los clientes al abrir sus cuentas, la información crítica de la campaña (como la duración de la llamada y el resultado del contacto) fue observada y registrada directamente por los operadores y los sistemas automáticos del call center.   
Por otro lado, los indicadores socioeconómicos incluidos en versiones extendidas del dataset fueron derivados de fuentes externas (como el banco central europeo) y acoplados a cada instancia según la fecha del contacto. La validación ‘y’ fue rigurosa: se verificó de forma determinística cruzando el registro de la llamada con el sistema transaccional del banco para confirmar si efectivamente se había dado de alta un depósito a plazo fijo.

**5\. Usos:**

**¿Se ha utilizado ya el conjunto de datos para alguna tarea? En caso afirmativo, descríbalo.**  
**¿Existe un repositorio que enlace a alguno o todos los documentos o sistemas que utilizan el conjunto de datos? En caso afirmativo, proporcione un enlace u otro punto de acceso.**

Sí, se ha utilizado en dataset en Marketing Predictivo, específicamente para aumentar la eficiencia de los call centers. El estudio original utilizó el dataset para reducir el número de llamadas necesarias. En lugar de llamar a todos los clientes, el modelo identifica al \~10% con mayor probabilidad de éxito, capturando el 50% de los depósitos totales. También se ha usado como "benchmark" para comparar la eficacia de modelos como Redes Neuronales, Support Vector Machines (SVM) y Random Forest. Para entender qué factores influyen más en la decisión del cliente. Por ejemplo, se descubrió que el éxito está fuertemente ligado a factores económicos externos (como las tasas de interés de la Eurozona) y a la duración de la llamada.  
Y por último, dado que la gran mayoría de los clientes dicen "no", investigadores lo usan para probar técnicas de balanceo de datos como SMOTE (Synthetic Minority Over-sampling Technique).
