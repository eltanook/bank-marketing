[^2]:  [https://www.sciencedirect.com/science/article/abs/pii/S016792361400061X?via%3Dihub](https://www.sciencedirect.com/science/article/abs/pii/S016792361400061X?via%3Dihub) (18/3/2026)














```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

df = pd.read_csv('data/full.csv', sep=';')

# Agrupar edad en etapas de vida
bins   = [17, 30, 64, 100]
labels = ['Joven (18-30)', 'Adulto (31-64)', 'Mayor (65+)']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Tasa de suscripcion por grupo etario
tasa = (df.groupby('age_group', observed=True)['y']
          .apply(lambda s: (s == 'yes').mean() * 100)
          .reset_index(name='pct_yes'))

conteo = df['age_group'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Grafico 1: cantidad de registros
axes[0].bar(conteo.index.astype(str), conteo.values, color=['#555555','#888888','#aaaaaa'])
axes[0].set_title('Cantidad de registros por grupo etario')
axes[0].set_ylabel('Registros')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Grafico 2: tasa de y=yes
axes[1].bar(tasa['age_group'].astype(str), tasa['pct_yes'], color=['#1A3A5C','#2E6DA4','#87B7DB'])
axes[1].set_title('Tasa de suscripcion (y=yes) por grupo etario')
axes[1].set_ylabel('% suscripcion')
axes[1].set_ylim(0, 35)
for i, v in enumerate(tasa['pct_yes']):
    axes[1].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('figures/ej1_age_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print('\nResumen:')
print(pd.concat([conteo.rename('N'), tasa.set_index('age_group')['pct_yes'].rename('% yes')], axis=1).to_string())
```

**v) Sesgo por edad (variable protegida oficial del dataset)**

El análisis de edad como variable protegida revela dos sesgos que se superponen y se refuerzan:

- **Subrepresentación del grupo Mayor (65+)**: representa una fracción pequeña del dataset. El modelo tiene pocas muestras de este grupo para aprender, lo que deteriora su capacidad predictiva para ese segmento.
- **Paradoja de la tasa base**: el grupo Mayor tiene la **tasa de suscripción más alta** (~27%) pero, al estar subrepresentado en el entrenamiento, el modelo tiende a ignorarlo. Esto genera **más Falsos Negativos** en ese grupo: personas con alta probabilidad de suscribirse que el modelo clasifica como 'no'.
- **Jóvenes** (18–30): también subrepresentados, aunque con tasa de 'yes' (~14%) similar al promedio.

> **Conclusión:** La `age` es una variable protegida clave en este dataset. El sesgo no es solo de representación sino de **tasa base diferencial**, lo que lo hace especialmente relevante desde la perspectiva de equidad.
